```python
from sklearn.datasets import fetch_20newsgroups
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Load the 20 newsgroups dataset
newsgroups = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))

# Extract the text from the documents
dataset = newsgroups.data[:5]  # Using the first 5 documents for illustration

stop_words = set(stopwords.words('english'))

filtered_dataset = []

for document in dataset:
    word_tokens = word_tokenize(document)
    filtered_words = [w for w in word_tokens if not w.lower() in stop_words]
    filtered_dataset.append(filtered_words)

print("Original dataset:")
for document in dataset:
    print(document)

print("\nFiltered dataset:")
for filtered_words in filtered_dataset:
    print(filtered_words)
```

```python
from nltk.corpus import movie_reviews
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the movie_reviews dataset if not already downloaded
import nltk
nltk.download('movie_reviews')

# Get the movie_reviews dataset from nltk
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]

# Take a few positive and negative reviews for demonstration
documents = documents[:5] + documents[-5:]

stop_words = set(stopwords.words('english'))

filtered_dataset = []

for words, category in documents:
    filtered_words = [w for w in words if not w.lower() in stop_words]
    filtered_dataset.append((filtered_words, category))

print("Original dataset:")
for words, category in documents:
    print(f"{category}: {' '.join(words)}")

```

```python
print("\nFiltered dataset:")
for filtered_words, category in filtered_dataset:
    print(f"{category}: {' '.join(filtered_words)}")
```
